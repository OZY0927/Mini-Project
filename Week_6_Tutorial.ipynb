```python
##Question 1
import cv2

# Load the image
image_path = 'images/boat.jpg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Otsu thresholding
_, thresholded = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
thresholded = cv2.bitwise_not(thresholded)
# Find contours in the thresholded image
contours, _ = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Find the largest contour (likely to be the boat and people)
largest_contour = max(contours, key=cv2.contourArea)

# Get the bounding rectangle of the largest contour
x, y, w, h = cv2.boundingRect(largest_contour)

# Draw a bounding box around the detected region
result = cv2.drawContours(image.copy(), [largest_contour], -1, (0, 255, 0), 2)
cv2.rectangle(result, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the result
cv2.imshow('Segmented Boat and People', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##Question2
import cv2
import numpy as np

# Function to update the displayed image based on trackbar value
def update_image(val):
    kernel_size = cv2.getTrackbarPos('Kernel Size', 'Morphological Transformations')
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))
    
    # Apply morphological transformations
    erosion = cv2.erode(thresholded, kernel)
    dilation = cv2.dilate(thresholded, kernel)
    opening = cv2.morphologyEx(thresholded, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(thresholded, cv2.MORPH_CLOSE, kernel)
    
    # Display images
    cv2.imshow('Erosion', erosion)
    cv2.imshow('Dilation', dilation)
    cv2.imshow('Opening', opening)
    cv2.imshow('Closing', closing)

# Load the image
image_path = 'images/leaf.jfif'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to segment the leaf
_, thresholded = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)

# Create a GUI window
cv2.namedWindow('Morphological Transformations')

# Create a trackbar to control the size of the structuring element
cv2.createTrackbar('Kernel Size', 'Morphological Transformations', 1, 20, update_image)

# Initial update to show the images
update_image(0)

# Wait for key press and close window when a key is pressed
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def color_seg(img, color1_low, color1_high, color2_low, color2_high, smooth=True):
    """Args:
    img: source img
    color_low, color_high: HSV color tuples
    smooth=Boolean (default:True)"""
    img_ori = img.copy()
    if smooth:
        img = cv2.GaussianBlur(img, (5, 5), 0)
        
    #HSV
    img_hsv = cv2.cvtColor(img, cv.COLOR_RGB2HSV)
    
    #inrange
    mask1 = cv2.inRange(img_hsv, color1_low, color1_high)
    mask2 = cv2.inRange(img_hsv, color2_low, color2_high)
    
    # Combine the masks
    mask = cv2.bitwise_or(mask1, mask2)
    
    #morphological operations
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations = 2)
    
    return cv2.bitwise_and(img_ori, img_ori, mask=mask)

imgs = []

for i in range(1,6):
    img = cv2.imread('images\color_space\porcelain'+str(i)+'.jfif')
    img = cv2.cvtColor(img,cv.COLOR_BGR2RGB)
    imgs.append(img)

res = [color_seg(img, light_blue_value, dark_blue_value, light_white_value, dark_white_value) for img in imgs]
```

```python
fig,ax = plt.subplots(5,2,figsize=(10,10))

for i in range(5):
    ax[i][0].imshow(imgs[i])
    ax[i][0].axis("off")
    ax[i][1].imshow(res[i])
    ax[i][1].axis("off")

fig.suptitle("Images and their segmentation")
plt.tight_layout()
plt.show()
```
